## b

```python
def kmeans(k, input_data):
    if 'cluster' in input_data.columns:
        input_data.drop('cluster', axis = 1, inplace = True)
    random_index = random.sample(range(0,X.shape[0]),k)
    centroids = input_data.iloc[random_index]
    centroids.reset_index(inplace=True, drop = True)
    for i in range(10):
        if 'cluster' in input_data.columns:
            input_data.drop('cluster', axis = 1, inplace = True)
        cluster_group = []
        distances = []
        for index, row in input_data.iterrows():
            for j in range(k):
                dist = np.linalg.norm(row - centroids.iloc[j])
                distances.append(dist)
            min_distance = min(distances)
            index_pos = distances.index(min_distance)
            cluster_group.append(index_pos)
            distances.clear()
        old_centroids = centroids
        new_centroids = pd.DataFrame(columns = centroids.columns)
        clusters = np.unique(cluster_group)
        input_data['cluster'] = cluster_group
        for cluster in clusters:
            X_without_cluster_label= input_data.loc[input_data['cluster'] == cluster]
            X_without_cluster_label.drop('cluster',axis = 1, inplace=True)
            new_centroids.at[cluster, :] = X_without_cluster_label.mean(axis=0)
            #new_centroids.append(X_without_cluster_label.mean(axis=0), ignore_index=True)
        centroids = new_centroids
        if new_centroids.equals(old_centroids) :
            print('Best Clusters found at the', i,'th iteration')
            break
    return input_data
```

```python
input_data = X.copy()
output_with_clusters = kmeans(3,input_data)
output_with_clusters.head()
```

```python
output_with_clusters['cluster'].value_counts()
```

#### In usual scenarios it is difficult to compare the labels in k means cluster to true labels. This is because the original label for first 50 rows might be class 1 but the cluster labels assigned through k means would be 2 for all of the 50 rows. This would mean that even though we are grouping them together and consistently but we cannot compare this with true labels 

#### The Adjusted Rand score is used compare the true labels with cluster labels which calculates the similarity between the two by taking all points identified within the same cluster.

```python
from sklearn.metrics import adjusted_rand_score
ari = adjusted_rand_score(Y, output_with_clusters['cluster'])
print("Adjusted Rand Index (ARI):", ari)
```

```python
def calculate_distance_matrix(data):
    n = data.shape[0]
    distance_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(i + 1, n):
            distance = np.linalg.norm(data[i]-data[j])
            distance_matrix[i, j] = distance_matrix[j, i] = distance
    return distance_matrix


def hierarchical_clustering(K, data):
    n = data.shape[0]
    distance_matrix = calculate_distance_matrix(data)
    clusters = [[i] for i in range(n)]

    while len(clusters) > K:
        min_distance = np.inf
        merge_indices = (0, 0)

        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                cluster1 = clusters[i]
                cluster2 = clusters[j]
                distance_sum = 0
                for c1 in cluster1:
                    for c2 in cluster2:
                        distance_sum += distance_matrix[c1, c2]
                avg_distance = distance_sum / (len(cluster1) * len(cluster2))
                if avg_distance < min_distance:
                    min_distance = avg_distance
                    merge_index = (i, j)
        i, j = merge_index
        clusters[i].extend(clusters[j])
        del clusters[j]

    cluster_labels = np.zeros(n)
    for i, cluster in enumerate(clusters):
        for data_point in cluster:
            cluster_labels[data_point] = i + 1

    return cluster_labels
```

```python
K = 3
hierarchical_labels = hierarchical_clustering(K, data)
accuracy = accuracy_score(np.array(df['label']), hierarchical_labels)
print("Accuracy:", round(accuracy,2))
```

```python
performance_metrics(df['label'], hierarchical_labels)
```

The performance is decreases by the barest of margins from 1 to 0.99. There is just 1 misclassification where label 1 is classified as label 0.

```python
pca = PCA(n_components=2)
principal_components  = pca.fit_transform(X_std)
X = pd.DataFrame(principal_components, columns  = ['pca_component_1','pca_component_2'])
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
kfold = skm.KFold(5,random_state=1,shuffle=True)
param_grid = {'C': [1,15], 
              'gamma': [1,0.1, 0.01],
              'kernel': ['rbf','poly'],
              'degree' : [1,2,3]} 
grid = GridSearchCV(SVC(), param_grid, cv = kfold, refit=True ,scoring = 'accuracy')
grid.fit(X_train, y_train)
grid.best_params_
```
